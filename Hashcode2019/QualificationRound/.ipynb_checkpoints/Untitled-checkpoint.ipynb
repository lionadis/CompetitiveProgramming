```python
import numpy as np
```

```python
class Slide:
    
    def __init__(self, tags, idx):
        
        self.idx = idx
        self.tags = tags
```

```python
class HalfSlide:
    
    def __init__(self, tags, idx):

        self.idx = idx
        self.tags = tags
```

```python
def parse_input(filename):
    
    data = open(filename).readlines()
    
    N = int(data[0])
    
    H, V = [], []
    
    data.pop(0)
    
    for i in range(len(data)):
        slide = data[i].split()
        if slide[0] == 'V':
            V.append(HalfSlide(slide[2:], i))
        else:
            H.append(Slide(slide[2:], i))
    return N, H, V
```

```python
N, H, V = parse_input('input/c_memorable_moments.txt')
```

```python
print(len(V))
```

```python
def all_slides(H, V):
    '''
        Randomly joining vertical slides
    '''
    slides = H.copy()
    for i in range(0, len(V), 2):
        slides.append(Slide(list(set(V[i].tags + V[i + 1].tags)), [V[i].idx, V[i + 1].idx]))
    return slides
```

```python
slides = all_slides(H, V)
```

```python
def f(s1, s2):
    
    return min(len(set(s1.tags) - set(s2.tags)), len(set(s2.tags) - set(s1.tags)), len(set(s1.tags) & set(s2.tags)))
```

```python
def score(slides):
    
    s = 0
    for i in range(0, len(slides) - 1, 2):
        s += f(slides[i], slides[i + 1])
```

```python
def output(slides, filename):
    f = open(filename, 'w+')
    print(len(slides), file=f)
    for s in slides:
        if isinstance(s.idx, list):
            print(s.idx[0], s.idx[1], file=f)
        else:
            print(s.idx, file=f)
```

```python
def greedy_1(_slides):
    '''
        append the slides which leads to the maximum score
    '''
    slides = _slides.copy()
    ans = [slides[0]]
    while len(ans) < len(slides):
        best_score, best_idx = -1, -1
        for i in range(len(slides)):
            _score =  f(ans[-1], slides[i])
            if best_score < _score:
                best_score = _score
                best_idx = i
        if best_score == 0:
            ans.extend(slides)
        else:
            ans.append(slides[best_idx])
            slides.pop(best_idx)
    return ans
```

```python
greedy_1(slides)
```
